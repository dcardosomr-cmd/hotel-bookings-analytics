#### 4. Limpeza dos dados

Remoção dos hóspedes "fantasma"

```python
df = df[~((df['adults'] == 0) & (df['children'] == 0) & (df['babies'] == 0))]
```

```python
df[df['adults'] + df['children'] + df['babies'] == 0]
```

Substituição dos valores a nulo

```python
df['agent'] = df['agent'].fillna(0)
df['company'] = df['company'].fillna(0)
df['children'] = df['children'].fillna(0)
df['country'] = df['country'].fillna('undefined')
```

Verificação da alteração

```python
df.isnull().sum()
```

#### 5. Correção dos data types

Conversão da data

```python
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'], errors='coerce')
```

Conversão dos valores númericos para inteiros

```python
df['children'] = df['children'].astype(int)
df['adults'] = df['adults'].astype(int)
df['babies'] = df['babies'].astype(int)
df['agent'] = df['agent'].astype(int)
df['company'] = df['company'].astype(int)
df['is_canceled'] = df['is_canceled'].astype(int)
df['is_repeated_guest'] = df['is_repeated_guest'].astype(int)
```

Conversão das strings

```python
categorical_cols = ['hotel',
                    'meal',
                    'country',
                    'market_segment',
                    'distribution_channel',
                    'reserved_room_type',
                    'assigned_room_type',
                    'deposit_type',
                    'customer_type',
                    'reservation_status']

df[categorical_cols] = df[categorical_cols].astype('category')
```

Verificação da alteração

#### 7. Resumo final

Após a importação dos dados originais, procedeu-se a uma análise da sua integridade para identificar incoerências, como hóspedes "fantasma" (adultos, crianças e bebés = 0), valores de ADR negativos e discrepâncias entre cancelamentos e o estado da reserva.

Os valores nulos foram tratados de acordo com o tipo de informação: nas variáveis numéricas, os nulos foram substituídos por 0, refletindo ausência da característica; na variável “country”, adotou-se o rótulo “undefined” para evitar perda de informação categórica.
